```python
# KROK 3 — silnik konwersji
from pdf2docx import Converter
import subprocess
import os

def pdf_to_docx(pdf_path, out_path):
    cv = Converter(pdf_path)
    cv.convert(out_path)
    cv.close()

def docx_to_pdf(docx_path):
    subprocess.run([
        "libreoffice",
        "--headless",
        "--convert-to",
        "pdf",
        docx_path,
        "--outdir",
        "/content"
    ])
```

```python
#KROK 4 — GUI (UPLOAD + PRZYCISKI)
from google.colab import files
from pdf2docx import Converter
import subprocess
import os

# =========================
# 🔧 KONWERSJE
# =========================

def pdf_to_docx(pdf_path, out_path):
    cv = Converter(pdf_path)
    cv.convert(out_path)
    cv.close()

def docx_to_pdf(docx_path):
    subprocess.run([
        "libreoffice",
        "--headless",
        "--convert-to",
        "pdf",
        docx_path,
        "--outdir",
        "/content"
    ])

# =========================
# 📤 UPLOAD (STABILNY)
# =========================

print("📤 Wrzuć plik (PDF lub DOCX):")
uploaded = files.upload()

for filename in uploaded.keys():
    file_path = "/content/" + filename

    with open(file_path, "wb") as f:
        f.write(uploaded[filename])

    print("\n📄 Wczytano:", filename)

    # =========================
    # 🔁 LOGIKA KONWERSJI
    # =========================

    if filename.lower().endswith(".pdf"):
        out_path = file_path.replace(".pdf", ".docx")
        pdf_to_docx(file_path, out_path)
        print("✅ PDF → DOCX gotowe:", out_path)

        files.download(out_path)

    elif filename.lower().endswith(".docx"):
        docx_to_pdf(file_path)
        out_pdf = file_path.replace(".docx", ".pdf")
        print("✅ DOCX → PDF gotowe:", out_pdf)

        files.download(out_pdf)

    else:
        print("❌ Nieobsługiwany format")
```
